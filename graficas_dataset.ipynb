```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating dataset for latency
np.random.seed(10)
data_aws = np.random.normal(21, 2.25, 300)
data_ubidots = np.random.normal(35, 5.25, 300)
data = [data_aws, data_ubidots]

print(data)
```

```python
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)

plt.xlabel('Servicios de almacenamiento de datos para IoT')
plt.ylabel('Latencia para carga de datos')
plt.title('Comparaci√≥n de latencia: AWS IoT vs Ubidots')

# save the figure
plt.savefig('plot.png', dpi=300, bbox_inches='tight')

# show plot
plt.show()
```
